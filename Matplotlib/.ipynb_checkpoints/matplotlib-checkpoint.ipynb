```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)     #it allows you to create evenly spaced arrays with a specified number of elements.  (start,end ,step))
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

### plot the data 

```python
#sin graph
plt.plot(x,y)
plt.show   
```

```python
#cos graph
plt.plot(x,z)
plt.show()
```

#### adding title x-axis y-axis labels

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine wave')
plt.show()
```

#### parabola

```python
x=np.linspace(-10,10,20)
y= x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(-10,10,20)
y= x**2
plt.plot(x,y,'r+')    # r is red color and + will show in the graph
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

#### Barplot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English', 'French',' Spanish', 'Latin', 'German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
plt.show()
```

#### pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English', 'French',' Spanish', 'Latin', 'German']
people=[100,50,150,40,70]
ax.pie(people, labels=languages, autopct='%1.1f%%' )
plt.show()
```

#### scatter plot

```python
x= np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

#### 3d scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
